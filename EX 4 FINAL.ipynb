```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return 100*(np.sin(x)-y)

def y_analytical(x):
    return -100*np.exp(-100*x)*(-100*np.exp(100*x)*np.sin(x)+np.exp(100*x)*np.cos(x)-1)/10001

def euler(f, a, b, n, yinit):
    h = (b-a)/(n-1)
    xs = np.linspace(a,b,n)
    ys = np.zeros(n)

    y = yinit
    for j,x in enumerate(xs):
        ys[j] = y
        y += h*f(x, y)
    return xs, ys

def rk4(f,a,b,n,yinit):
    h = (b-a)/(n-1)
    xs = np.linspace(a,b,n)
    ys = np.zeros(n)

    y = yinit
    for j,x in enumerate(xs):
        ys[j] = y
        k0 = h*f(x, y)
        k1 = h*f(x+h/2, y+k0/2)
        k2 = h*f(x+h/2, y+k1/2)
        k3 = h*f(x+h, y+k2)
        y += (k0 + 2*k1 + 2*k2 + k3)/6
    return xs, ys

if __name__ == '__main__':
    a, b, n, yinit = 1, 3, 35, 20
    xs, ys_euler = euler(f,a,b,n,yinit)
    xs, ys_rk4 = rk4(f,a,b,n,yinit)
    x_ana = np.linspace(a,b,100)
    y_ana = y_analytical(x_ana)
    #plt.plot(xs,ys_euler,'bs',label='euler')
    plt.plot(xs,ys_rk4)
    plt.plot(x_ana,y_ana)
```
