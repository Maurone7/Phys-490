```python
import numpy as np
import matplotlib.pyplot as plt

x_i = np.arange(0, 1.68, 0.08)
```

```python
def function(x):
    return np.e**(np.sin(2*x))

plt.plot(x_i, function(x_i))
plt.show()
```

```python
def function_prime(x, h):
    return (function(x + h) - function(x))/ h

def function_derivative(x):
    return 2*np.e**(np.sin(2*x))*np.cos(2*x)

plt.plot(x_i, function_prime(x_i - 0.04, 0.04), label='Central Approximation')
plt.plot(x_i, function_prime(x_i, 0.08), label='Forward Approximation')
plt.plot(x_i, function_derivative(x_i), label='Derivative')
plt.legend()
plt.show()
```
